```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts


data = pd.read_csv('insurance.csv')
#Ищем среднее, моду, медиану BMI
mean_bmi = np.mean(data["bmi"])
moda_bmi = sts.mode(data["bmi"])
median_bmi = np.median(data["bmi"])
#Ищем среднее, моду, медиану Charges
mean_charges = np.mean(data["charges"])
moda_charges = sts.mode(data["charges"])
median_charges = np.median(data["charges"])
#Ищем стандартное отклонение, минимум, максимум, размах BMI
std_bmi = np.std(data["bmi"])
min_bmi = np.min(data["bmi"])
max_bmi = np.max(data["bmi"])
range_bmi = max_bmi - min_bmi
#Ищем стандартное отклонение, минимум, максимум, размах Charges
std_charges = np.std(data["charges"])
min_charges = np.min(data["charges"])
max_charges = np.max(data["charges"])
range_charges = max_charges - min_charges
#Выводим значения текстом BMI
print("BMI:")
print("Mean:", mean_bmi)
print("Median:", median_bmi)
print("Mode:", moda_bmi.mode[0])
print("Standard Deviation:", std_bmi)
print("Minimum:", min_bmi)
print("Maximum:", max_bmi)
print("Range:", range_bmi)
#Выводим значения текстом Charges
print("charges:")
print("Mean:", mean_charges)
print("Median:", median_charges)
print("Mode:", moda_charges.mode[0])
print("Standard Deviation:", std_charges)
print("Minimum:", min_charges)
print("Maximum:", max_charges)
print("Range:", range_charges)

#Выводим значения на гистограммах BMI
plt.subplot(2, 1, 1)
plt.hist(data["bmi"], bins=30, color="skyblue")
plt.axvline(mean_bmi, color="red", linestyle="dashed", linewidth=2, label="Mean")
plt.axvline(median_bmi, color="green", linestyle="dashed", linewidth=2, label="Median")
plt.axvline(moda_bmi.mode[0], color="orange", linestyle="dashed", linewidth=2, label="Mode")
plt.axvline(range_bmi, color="blue", linestyle="dashed", linewidth=2, label="Range")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.legend()
#Выводим значения на гистограммах Charges
plt.subplot(2, 1, 2)
plt.hist(data["charges"], bins=30, color="skyblue")
plt.axvline(mean_charges, color="yellow", linestyle="dashdot", linewidth=2, label="Mean")
plt.axvline(median_charges, color="purple", linestyle="dashdot", linewidth=2, label="Median")
plt.axvline(moda_charges.mode[0], color="black", linestyle="dashdot", linewidth=2, label="Mode")
plt.axvline(range_charges, color="pink", linestyle="dashdot", linewidth=2, label="Range")
plt.xlabel("charges")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.show
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('insurance.csv')
#Создадим список, содержащий 300 выборок длиной по n элементов
bmi_samples = [data['bmi'].sample(40) for _ in range(300)]
#Вычислим среднее и стандартное отклонение для каждой выборки
means = [sample.mean() for sample in bmi_samples]
stds = [sample.std() for sample in bmi_samples]
#Выведем гистограммы для средних значений и стандартных отклонений
plt.hist(means, bins=30)
plt.xlabel('Среднее значение BMI')
plt.ylabel('Частота')
plt.title('Гистограмма средних значений BMI')
plt.show()

plt.hist(stds, bins=30)
plt.xlabel('Стандартное отклонение BMI')
plt.ylabel('Частота')
plt.title('Гистограмма стандартных отклонений BMI')
plt.show()

print('Среднее значение BMI:', np.mean(means))
print('Стандартное отклонение BMI:', np.std(stds))
```

```python
import pandas as pd
import numpy as np
from scipy import stats

data = pd.read_csv("insurance.csv")
# Вычисление выборочного среднего и стандартного отклонения для charges и bmi
mean_charges = data["charges"].mean()
std_charges = data["charges"].std()

mean_bmi = data["bmi"].mean()
std_bmi = data["bmi"].std()

# Расчет доверительных интервалов для charges и bmi на уровне доверия 95%
n = len(data)
confidence_level = 0.95
Z = stats.norm.ppf(confidence_level) #критическое значение

charges_interval_95 = (mean_charges - Z * std_charges / np.sqrt(n), mean_charges + Z * std_charges / np.sqrt(n))
bmi_interval_95 = (mean_bmi - Z * std_bmi / np.sqrt(n), mean_bmi + Z * std_bmi / np.sqrt(n))
# Расчет доверительных интервалов для charges и bmi на уровне доверия 99%
confidence_level = 0.99
Z = stats.norm.ppf(confidence_level)

charges_interval_99 = (mean_charges - Z * std_charges / np.sqrt(n), mean_charges + Z * std_charges / np.sqrt(n))
bmi_interval_99 = (mean_bmi - Z * std_bmi / np.sqrt(n), mean_bmi + Z * std_bmi / np.sqrt(n))

print("Доверительный интервал для среднего значения charges (95% уровень доверия):", charges_interval_95)
print("Доверительный интервал для среднего значения bmi (95% уровень доверия):", bmi_interval_95)

print("Доверительный интервал для среднего значения charges (99% уровень доверия):", charges_interval_99)
print("Доверительный интервал для среднего значения bmi (99% уровень доверия):", bmi_interval_99)
```

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

data = pd.read_csv('insurance.csv')
bmi_data = data['bmi']
charges_data = data['charges']

# Проводим KS-тест для признака bmi
ks_statistic, p_value = stats.kstest(bmi_data, 'norm')
print("KS-тест для признака bmi:")
print("KS-статистика:", ks_statistic)
print("p-значение:", p_value)

# Строим Q-Q plot для признака bmi
stats.probplot(bmi_data, dist='norm', plot=plt)
plt.title('Q-Q plot для признака bmi')
plt.show()

# Проводим KS-тест для признака charges
ks_statistic, p_value = stats.kstest(charges_data, 'norm')
print("KS-тест для признака charges:")
print("KS-статистика:", ks_statistic)
print("p-значение:", p_value)

# Строим Q-Q plot для признака charges
stats.probplot(charges_data, dist='norm', plot=plt)
plt.title('Q-Q plot для признака charges')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts


data = pd.read_csv('ECDCCases.csv')
data.describe()
# Проверка наличия пропущенных значений
missing_values = data.isnull().sum()
total_missing = missing_values.sum()
# Вывод количества пропущенных значений в процентах
percent_missing = (total_missing / (data.shape[0] * data.shape[1])) * 100 
#shape - кортеж размеров массива
#Форма массива - кол-во элементов в каждом измерении
print(f"Количество пропущенных значений:({percent_missing}%)")
# Удаление двух признаков с наибольшим количеством пропущенных значений
data = data.dropna(axis=1, thresh=data.shape[0] - 2) 
#dropna для удаления пустых значений

# Обработка пропусков для оставшихся признаков
for column in data.columns:
    if data[column].dtype == "object":
        data[column] = data[column].fillna("other")
    else:
        median_value = data[column].median()
        data[column] = data[column].fillna(median_value)
        
# Проверка отсутствия пропусков в данных
missing_values = data.isnull().sum().sum()
print(f"А есть ли пропущенные значения сейчас?: {missing_values}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts


data = pd.read_csv('ECDCCases.csv')
stat = data.describe()
# Поиск выбросов (значений, выходящих за пределы интерквартильного размаха) для числовых признаков
out_val = {} #словарь
for column in data.columns:
    if data[column].dtype != "object":
        IQR = stat[column]["75%"] - stat[column]["25%"]
        lower_bound = stat[column]["25%"] - (1.5 * IQR)
        upper_bound = stat[column]["75%"] + (1.5 * IQR)
        column_out = data[(data[column] < lower_bound) | (data[column] > upper_bound)].shape[0]
        out_val[column] = column_out

# Вывод признаков, содержащих выбросы и количество выбросов для каждого признака
print("Признаки, содержащие выбросы:")
for column, count in out_val.items():
    print(f"{column}: {count} выбросов")

# Поиск стран, для которых количество смертей в день превысило 3000 и количество таких дней
death_countries = data[data["deaths"] > 3000]["countryterritoryCode"].unique()
death_days = data[data["deaths"] > 3000].shape[0]
print(f"Количество стран, где количество смертей в день превысило 3000: {len(death_countries)}")
print(f"Количество дней, когда количество смертей в день превысило 3000: {death_days}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats import weightstats as stests


data = pd.read_csv('bmi.csv')
# Отбор выборок по регионам
northwest = data[data['region'] == 'northwest']['bmi']
southwest = data[data['region'] == 'southwest']['bmi']
# Проверка выборок на нормальность критерий ШопироУилка
shapiro_test_northwest = stats.shapiro(northwest)
shapiro_test_southwest = stats.shapiro(southwest)
print("Проверка на нормальность Шапиро-Уилка для выборки northwest:", shapiro_test_northwest)
print("Проверка на нормальность Шапиро-Уилка для выборки southwest:", shapiro_test_southwest)
# Проверка выборок на гомогенность дисперсии Критерий Бартлетта
bartlett_test = stats.bartlett(northwest, southwest)
print("Проверка на гомогенность дисперсии Тест Бартлетта:", bartlett_test)
# Проведение t-критерия Стьюдента
t_test = stests.ttest_ind(northwest, southwest, alternative='two-sided')
print("t-критерий Стьюдента:", t_test)
```

```python
import scipy.stats

observed = [97, 98, 109, 95, 97, 104]
expected = [100, 100, 100, 100, 100, 100]  # ожидаемое распределение, если кубик равномерный
#Ожидаемое распределение было найдено по формуле математического ожидания
#(среднее значение случаной величины), т.е 1/6*(97+98+109+95+97+104)
chi2, p = scipy.stats.chisquare(observed, expected)

print("Статистика Хи-квадрат:", chi2)
print("p-значение:", p)
```
